```python
class FoodItem():
    def __init__(self,name = None ,quantity = None ,price = 0 ,discount = 0,stock = 0):
        
        self.name = name
        
        self.quantity = quantity 
        
 #LOGIC for getting the correct input from the user or else values will  initate to zero       
        #checking value is int or not
        if type(price) != int:
            #checking value contains keyword none
            if price.upper() != 'NONE':
                #checking value is of only digits
                if price.isdigit() :
                    #typecasting value to int
                    self.price = int(price)
                else:
                    print('Invalid Price Input!initializing price value to 0')
                    self.price = 0
            else:
                self.price = 0
        else:
            self.price = price
                    
        
        if  type(discount) != int:
            if discount.upper() != 'NONE':
                if discount.isdigit():
                    self.discount = int(discount)
                else:
                    print('Invalid Discount Input! initializing discount value to 0')
                    self.discount = 0
            else:
                self.discount = 0
        else:
            self.discount = discount
                    
        if type(stock) != int:
            if stock.upper() != 'NONE':
                    if stock.isdigit():
                        self.stock = int(stock)
                    else:
                        print('Invalid Stock Input! initializing stock value to 0')
                        self.stock = 0
            else:
                self.stock = 0
        else:
            self.stock = stock


        self.bill_price = self.price * (1 - self.discount/100)   #this line is not used anywhere

#For displaying the details of food_items        
    def display_item(self):
        my_item = [self.name,self.quantity,self.price,self.discount,self.stock]
        return my_item
    
#For Editing food_items 
    def edit_item(self,name = None, quantity = None, price = 0, discount = 0,stock = 0):
        if  name.upper() != 'NONE':
            self.name = name
            
        if  quantity.upper() != 'NONE':
            self.quantity = quantity
            
        if price.upper() != 'NONE':
            if type(price) != int:
                if price.isdigit() :
                    self.price = int(price)
                else:
                    print('Invalid Price Input! initializing price to 0')
                    self.price = 0
            else:
                self.price = price
                    
        if discount.upper() != 'NONE':
            if  type(discount) != int:
                if discount.isdigit():
                    self.discount = int(discount)
                else:
                    print('Invalid Discount Input! initializing discount to 0')
                    self.discount = 0
            else:
                self.discount = discount
                    
        if stock.upper() != 'NONE':
            if type(stock) != int:
                if stock.isdigit():
                    self.stock = int(stock)
                else:
                    print('Invalid Stock Input initializing stock to 0')
                    self.stock = 0
            else:
                self.stock = stock
        self.bill_price = self.price * (1 - self.discount/100)
#for deleting food items           
    def __del__(self):
        pass
#         print(f'Item {self.name} successfully deleted')(intentionally commented out)
        
        
    def user_ordered(self):
        if self.stock:
            self.stock-=1
#             print(f'{self.name} ordered successfully')(intentionally commented out)
        else:
            pass
#             print(f'sorry {self.name} item is out of stock')(intentionally commented out)
```

## Random ID generator functions

```python
#Funciton to generate food_id for food_items

def food_id_generator():
    global my_food_id
    global current_id
    current_id = 'FID' + ''.join([random.choice(string.digits) for n in range(7)])
    print('your current food id for the item is: ',current_id)
    return current_id
```

```python
#Function to generate order_id for order_items
def generate_order_id():
    t_n = datetime.now()
    order_id = 'OID'+str(datetime.date(t_n)).replace('-','')+str(t_n.hour)+str(t_n.minute)+str(t_n.second)
    return order_id
```

```python
#Function for entering a name

def input_user_name():
    
    Name = None
    
    while not Name :
        Name = input('Enter your name here:')
        k = Name.split()
        temp = 0
        for i in Name.split():
            if not i.isalpha():
                temp = 1
                print('Please enter a valid name: "only enter alphabets, avoid numbers and full stop"')
                break
        if temp == 0:
            clear_output()
            print('Your name is: ',Name)
            break
        else:
            Name = None
    return Name
        
```

```python
#Function for entering a valid mobile number
def input_user_phone_number():
    
    Number = None
    
    while not Number :
        Number = input('Enter your Phone Number here:')
        pattern = '^\d{10}$'
        if len(re.findall(pattern,Number)):
            clear_output()
            print('Your phone number is: ',Number)
            break
        else:
            print('Enter a Valid Phone Number:')
            Number = None
            
    return Number
        
```

```python
#Function for entering valid email_id
def input_user_email():
    
    Email = None
    
    while not Email :
        Email = input('Enter your Email Id:')
        pattern = r'^[a-zA-Z0-9+_.-]+@[a-zA-Z0-9.-]+[a-z]+$'
        if len(re.findall(pattern,Email)):
            clear_output()
            print('Your email is: ',Email)
            break
        else:
            print('Enter a Valid Email:')
            Email = None
    return Email
```

```python
def adding_food_item():
    
    #defining global variables
    
    global my_food_id, my_item_list_objects, my_id_item_pair
    global my_id_item_pair_details, my_id_item_pair_complete_details 
    
    #here

    #Generating food_id using food_id_generator function( refer above function)
    adding = True
    while adding:

        current_food_id = food_id_generator()

        #Although Id might be random still there is a chance of 1 in million for id to be a match in existing food items
        #so using while loop, incase there is a match app starts from begining 

        while current_food_id not in my_food_id:

            #adding food id to our food_id list to keep it in check
            try:
                my_food_id.append(current_food_id)

                my_item = food_item_input()

                    #appending FoodItem object to my_item_list_objects

                my_item_list_objects.append(my_item)

                #updating our id-object dictionary

                my_id_item_pair[current_food_id] = my_item
                
            except:
                my_food_id.pop()
                print('food item not added something went wrong')

            #updating  other dictionaries for identifying  

            updating_dictionaries()
            print('\033[1mFoodItem added successfully')
            break
        try:
            re_add = input('Do you want to add another food item[Yes/No]\033[0;0m')

            if re_add.lower()[0] == 'y':
                clear_output()
                continue
            else:
                clear_output()
                adding = False
        except:
            clear_output()
            adding = False
            pass
```
